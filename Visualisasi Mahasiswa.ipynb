```python
import pandas as pd

file_path = "Data_Mahasiswa.csv"  # Sesuaikan dengan lokasi file
df = pd.read_csv(file_path, sep=";", engine="python")

print(df.head())  # Menampilkan 5 baris pertama
print(df.columns)  # Melihat nama kolom yang tersedia
```

```python
import matplotlib.pyplot as plt
angkatan_col = "Angkatan Mahasiswa"
skill_col = "Programming Skill"
df_grouped = df.groupby(angkatan_col)[skill_col].mean()

plt.figure(figsize=(10, 5))
plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-", color="b", label="Rata-rata Programming Skill")

plt.xlabel("Angkatan Mahasiswa")
plt.ylabel("Rata-rata Programming Skill")
plt.title("Tren Programming Skill Berdasarkan Angkatan Mahasiswa")
plt.legend()
plt.grid(True)

plt.show()
```
