```python
import pandas as pd

# Load dataset
file_path = "student-data.csv"  
df = pd.read_csv(file_path, sep=";")
df_grouped = df.groupby("age")["absences"].mean()
```

```python
import matplotlib.pyplot as plt

# Plot line chart
plt.figure(figsize=(10, 5))
plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-", color="b", label="Rata-rata Absensi")

# Tambahkan label dan judul
plt.xlabel("Usia Mahasiswa")
plt.ylabel("Rata-rata Absensi")
plt.title("Tren Absensi Berdasarkan Usia Mahasiswa")
plt.legend()
plt.grid(True)

# Tampilkan plot
plt.show()
```
